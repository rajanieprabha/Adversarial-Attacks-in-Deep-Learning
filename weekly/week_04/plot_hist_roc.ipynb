```python
import random
import numpy as np
from matplotlib import pyplot as plt
from sklearn import metrics
random.seed(0)
```

```python
def plot_entropy_roc(plot_dict, save=False):
    n = plot_dict["n"]
    adv_n = plot_dict["adv_n"]
    adv_succ = plot_dict["adv_succ"]
    noise_n = plot_dict["adv_n"]
    
    
    # generate data for given parameters
    x = [random.gauss(plot_dict["mu"],plot_dict["sd"]) for _ in range(plot_dict["n"])]
    x_noise = [random.gauss(plot_dict["noise_mu"],plot_dict["noise_sd"]) for _ in range(noise_n)]
    x_adv = [random.gauss(plot_dict["adv_mu"],plot_dict["adv_sd"]) for _ in range(adv_n)]
    
    scores = x+x_noise+x_adv
    
    tp = int(adv_n*adv_succ)
    fn = (adv_n-tp)
    labels = n*[0] + noise_n*[0] + tp*[1] + fn*[0]
    fpr, tpr, thresholds = metrics.roc_curve(labels, scores)

    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
    bins = np.linspace(0, plot_dict["linsapce_up"], 100)
    
    ax1.hist(x, bins, alpha=0.5, label='x_test')
    ax1.hist(x_noise, bins, alpha=0.5, label='x_plus_noise')
    ax1.hist(x_adv, bins, alpha=0.5, label='x_adv')
    ax1.legend(loc='upper right')
    ax1.set_xlabel("uncertainty (entropy)")

    lw = 2
    ax2.plot(fpr, tpr, color='darkorange', lw=lw)
    ax2.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    ax2.set_xlim([0.0, 1.0])
    ax2.set_ylim([0.0, 1.05])
    ax2.set_xlabel('False Positive Rate')
    ax2.set_ylabel('True Positive Rate')
    ax2.set_title('ROC')
    ax2.legend(loc="lower right")

    f.tight_layout()

    if save:
        plt.savefig(plot_dict["figname"]+'.png')
    else:
        plt.show()
```
